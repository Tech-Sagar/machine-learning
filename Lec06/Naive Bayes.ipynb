```python
import quandl
```

```python
df = quandl.get('WIKI/GOOGL')
len(df)
```

```python
target = df['Adj. High'][5:]
print(len(target))
features = df[:3225]
len(features)
```

```python
del features['Open']
del features['High']
del features['Low']
del features['Close']
del features['Volume']
del features['Ex-Dividend']
del features['Split Ratio']
del features['Adj. High']
```

```python
features
```

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
x_train = pd.DataFrame(features[:2902])
y_train = pd.DataFrame(target[:2902])
x_test = pd.DataFrame(features[2902:])
y_test = pd.DataFrame(target[2902:])
```

```python
reg.fit(x_train, y_train)
reg.score(x_test, y_test)
```
